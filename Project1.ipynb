```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
# Load the Iris dataset
data = load_iris()

# Convert to a Pandas DataFrame
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target

# Preview the data
df.head()
```

```python
#Data Cleaning
# Check for missing values
df.isnull().sum()
```

```python
#Optional:If there are missing values, decide how to handle them. You can either fill them with the mean, median, or mode, or drop rows/columns with missing values.
# Example of filling missing values with the mean of the column
# df.fillna(df.mean(), inplace=True)
```

```python
# Check for duplicates
df.duplicated().sum()
```

```python
# Remove duplicates if any
df = df.drop_duplicates()
```

```python
#Basic Statistical Analysis
# Statistical summary
df.describe()
```

```python
#Relationsgips
# Correlation matrix
df.corr()
```

```python
#Data Visualization
# 1.Histogram: To see the distribution of each feature.

# Plot histograms for each feature
df.hist(bins=10, figsize=(10, 8))
plt.tight_layout()
plt.show()

```

```python
# 2. Boxplot: To check for outliers.
# Boxplot for each feature
plt.figure(figsize=(10, 6))
sns.boxplot(data=df.drop('target', axis=1))
plt.show()
```

```python
# 3. Pairplot: To visualize the relationships between features and the target variable.

# Pairplot to visualize feature relationships
sns.pairplot(df, hue='target', palette='Set2')
plt.show()
```

```python
# 4.Heatmap: To visualize the correlation matrix.
# Correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.show()
```
